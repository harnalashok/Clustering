```python
# 1.0
%reset -f
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture as gmm
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
```

```python
# 1.1
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# 2.0
X, y = make_moons(200, noise=.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1]);
```

```python
# 2.1
kmeans = KMeans(5, random_state=0)
labels = kmeans.fit(X).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,  cmap='viridis');
```

```python
# 3.0
gm = gmm(n_components=6)     # Try with 2,3,4,5,6. Six gives good results
labels = gm.fit(X).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,  cmap='viridis');
```

```python
# 4.0 gmm as generator
gm.sample(n_samples = 5)
```

```python
# 4.1 Generate 300 samples
y = gm.sample(n_samples = 300)
```

```python
# 4.2 Same kind of samples are generated 
s = y[0]
clr = y[1]
plt.scatter(s[:,0], s[:,1], c = clr) ;
```
